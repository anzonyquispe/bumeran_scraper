```python
#Library
# ---
# For scrape
from selenium import webdriver
from webdriver_manager.chrome import ChromeDriverManager

# Options driver
from selenium.webdriver.chrome.options import Options

# For dataframes
import pandas as pd
from pandas import json_normalize

# For simulate human behavior.
import time
from time import sleep
import random

# Clear data
import unidecode
# unaccented_string = unidecode.unidecode(accented_string)

# Manipulate Json files
import json
```

```python
# Setting
options= Options()
options.add_argument("--start-maximized")

driver = webdriver.Chrome(ChromeDriverManager().install(),options=options)
```

```python
# we open our page
driver.get("https://www.bumeran.com.pe/")
```

```python
# Puestos_ejecutivo

# //*[@id="root"]/div/div[1]/div/div/div/div/div/div/ul/li[5]/div/a     a esto   //div[@cless="sc-eNQAEJ fAbPPh"


driver.find_element_by_xpath('//div[@class="sc-eNQAEJ fAbPPh"]').click()
```

```python
listado_de_avisos = driver.find_element_by_id('listado-avisos')
```

```python
urls = []
links = listado_de_avisos.find_elements_by_tag_name('a')
for link in links[2:]:
    urls.append(link.get_attribute('href'))
```

```python
# The main dictionary
all_info = {}
all_info['executive_jobs']=[]



for url in urls:
    driver.get(url)
    time.sleep(random.randint(3,5))
    
    page = {}

# We find elements (key-value) and add them to the main dictionary

    # Location by xpath
    puesto = driver.find_element_by_xpath('//h1[contains(@class,"Title__H1-sc-2yd2j1-0 dmYIZt")]').text
    # Append
    page['PUESTO'] = puesto
    
    # Select icons clock
    icons_clock = driver.find_elements_by_xpath('//i[@name="icon-light-clock"]/following-sibling::h2')
    
#                                    SECTION RED

    # Location by index of icons_clock
    publication_date = icons_clock[0].text
    # Append
    page['PUBLICATION_DATE'] = publication_date

    # Location by xpath using name
    location = driver.find_element_by_xpath('//i[@name="icon-light-location-pin"]/following-sibling::h2').text
    # Append
    page['LOCATION'] = location
    
    try:
        # Location by xpath using name
        disability = driver.find_element_by_xpath('//a[@href="/empleos-discapacitados-apto.html"]').text
        # Append
        page['SUITABLE_FOR_PEOPLE_WITH_DISABILITIES'] = disability
    except:
        pass
#                                    SECTION BLUE

    # Location by index of icons_clock
    type_of_schedule = icons_clock[1].text
    # Append
    page['SCHEDULES_AVAILABLE'] = type_of_schedule

    # Location by xpath using name
    money = driver.find_element_by_xpath('//i[@name="icon-light-money"]/following-sibling::h2').text
    # Append
    page['SALARY'] = money

    # Location by xpath using name
    area = driver.find_element_by_xpath('//i[@name="icon-light-cube"]/following-sibling::h2').text
    # Append
    page['AREA'] = area


    # Location by xpath using name
    modality = driver.find_element_by_xpath('//i[@name="icon-light-office"]/following-sibling::h2').text
    # Append
    page['MODALITY_AVAILABLE'] = modality



    # Location by xpath
    details = driver.find_element_by_xpath('//div[@class="sc-hARARD ewkLMq"]').text
    details = unidecode.unidecode(details).replace('\n',' ')
    # Append
    page['DETAILS_AND_REQUIREMENTS'] = details
    
    all_info['executive_jobs'].append(page)
    
```

```python
data = json_normalize(all_info['executive_jobs'])
data
```

```python
# Export df to excel
data.to_excel('Bumeran_jobs.xlsx')
```

```python
# Export df to Json
data.to_json('Bumeran_jobs.json')
```
